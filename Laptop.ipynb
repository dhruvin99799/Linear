```python
# str.findall(r'(\d+\.?\d+)'): This applies a regular expression pattern r'(\d+\.?\d+)' 
# to each element in the selected column using the .str.findall() method. The regular 
# expression pattern matches any sequence of one or more digits \d+, followed by an 
# optional decimal point \.?, followed by zero or more digits \d*. The parentheses 
# () are used to capture the matched sequence as a group.
df['X_res'] = df['X_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x:x[0])
```

```python
X=df[['Company','TypeName','Inches','Ram','Weight','Touchscreen','IPS','ppi','Intel Core i5','Intel Core i3','Intel Core i7','Amd proceeser','Other Intel proceeser','Gpu brand','Mac','Others/No OS/Linux','Windows','HDD','SSD','FlashStorage']]
Y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=35)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
oho = OneHotEncoder()
oho.fit(X[['Company','TypeName','Inches','Ram','Weight','Touchscreen','IPS','ppi','Intel Core i5','Intel Core i3','Intel Core i7','Amd proceeser','Other Intel proceeser','Gpu brand','Mac','Others/No OS/Linux','Windows','HDD','SSD','FlashStorage']])
```

```python
col_trans = make_column_transformer((OneHotEncoder(categories=oho.categories_),['Company','TypeName','Inches','Ram','Weight','Touchscreen','IPS','ppi','Intel Core i5','Intel Core i3','Intel Core i7','Amd proceeser','Other Intel proceeser','Gpu brand','Mac','Others/No OS/Linux','Windows','HDD','SSD','FlashStorage']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(col_trans,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)*100
```
